```python
#加载 python 库
import numpy as np
import cv2
# 输入路径读取图片
img = cv2.imread('cat.jpg')
#定义窗口名字，和需要输出的图片
cv2.imshow('OpenCV CAT',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
